```python
import pickle
import scipy as sp
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

sns.set()
```

## Data Ingestion and Preparation

```python
df_purchase = pd.read_csv("data\purchase data.csv", index_col=0)

scaler = pickle.load(open('scaler.pickle', 'rb'))
pca = pickle.load(open('pca.pickle', 'rb'))
kmeans_pca = pickle.load(open('kmeans_pca.pickle', 'rb'))

features = df_purchase[['Sex', 'Marital status', 'Age', 'Education', 'Income', 'Occupation', 'Settlement size']]
df_purchase_segm_std = scaler.transform(features)

df_purchase_segm_pca = pca.transform(df_purchase_segm_std)

purchase_segm_kmeans_pca = kmeans_pca.predict(df_purchase_segm_pca)

df_purchase_predictors = df_purchase.copy()
df_purchase_predictors['Segment'] = purchase_segm_kmeans_pca

segment_dummies = pd.get_dummies(purchase_segm_kmeans_pca, prefix='Segment', prefix_sep = '_')
df_purchase_predictors.join(segment_dummies)

df_pa = df_purchase_predictors
```

## Purchase Probability Model

```python
y = df_pa['Incidence']
X = pd.DataFrame()
X['Mean Price'] = (df_pa['Price_1'] + 
                  df_pa['Price_2'] + 
                  df_pa['Price_3'] + 
                  df_pa['Price_4'] + 
                  df_pa['Price_5'])/5
```

```python
model_purchase = LogisticRegression(solver='sag')
model_purchase.fit(X,y)
```

```python
model_purchase.coef_
```

## Price Elasticity of Purchase Probability

```python
df_pa[['Price_1', 'Price_2', 'Price_3', 'Price_4', 'Price_5']].describe()
```

```python
price_range = np.arange(0.5, 3.5, 0.1)
price_range
```

```python
df_price_change = pd.DataFrame(price_range, columns=['Mean Price'])
```

```python
y_pr = model_purchase.predict_proba(df_price_change)
purchase_pr = y_pr[:,1]
pe = model_purchase.coef_[:, 0] * price_range * (1-purchase_pr)
```

```python
df_price_elasticities = pd.DataFrame(price_range)
```

```python
df_price_elasticities = df_price_elasticities.rename(columns={0: 'Price_Point'})
df_price_elasticities['Mean PE'] = pe
df_price_elasticities
```

```python
plt.figure(figsize=(12, 9))
plt.plot(price_range, pe, color='grey')
plt.xlabel('Price')
plt.ylabel('Elasticity')
plt.title('Price Elasticity of Purchase Probability');
```

## Purchase Probability with Promotion Feature

### Data Preparation

```python
y = df_pa['Incidence']

X = pd.DataFrame()
X['Mean Price'] = (df_pa['Price_1'] + 
                  df_pa['Price_2'] + 
                  df_pa['Price_3'] + 
                  df_pa['Price_4'] + 
                  df_pa['Price_5'])/5
```

```python
X['Mean_Promotion'] = (df_pa['Promotion_1'] +
                      df_pa['Promotion_2'] +
                      df_pa['Promotion_3'] +
                      df_pa['Promotion_4'] +
                      df_pa['Promotion_5'])/5
```

### Model Estimation

```python
model_incidence_promotion = LogisticRegression(solver='sag')
model_incidence_promotion.fit(X, y)
model_incidence_promotion.coef_
```

### Price Elasticity with Promotion

```python
df_price_elasticity_promotion = pd.DataFrame(price_range)
df_price_elasticity_promotion = df_price_elasticity_promotion.rename(columns = {0: 'Price Range'})

df_price_elasticity_promotion['Promotion'] = 1
y_promotion = model_incidence_promotion.predict_proba(df_price_elasticity_promotion.rename(columns={'Price Range': 'Mean Price',
                                                                                                   'Promotion': 'Mean_Promotion'}))

promo = y_promotion[:, 1]
price_elasticity_promo = (model_incidence_promotion.coef_[:, 0]*price_range) * (1-promo)

df_price_elasticities['Elasticity_Promo_1'] = price_elasticity_promo
df_price_elasticities.head()
```

### Price Elasticity without Promotion

```python
df_price_elasticity_no_promotion = pd.DataFrame(price_range)
df_price_elasticity_no_promotion = df_price_elasticity_no_promotion.rename(columns = {0: 'Price Range'})

df_price_elasticity_no_promotion['Promotion'] = 0
y_no_promotion = model_incidence_promotion.predict_proba(df_price_elasticity_no_promotion.rename(columns={'Price Range': 'Mean Price',
                                                                                                   'Promotion': 'Mean_Promotion'}))

promo = y_no_promotion[:, 1]
price_elasticity_no_promo = (model_incidence_promotion.coef_[:, 0]*price_range) * (1-promo)

df_price_elasticities['Elasticity_Promo_0'] = price_elasticity_no_promo
df_price_elasticities.head()
```

```python
plt.figure(figsize=(12, 9))
plt.plot(price_range, price_elasticity_promo, color='grey')
plt.plot(price_range, price_elasticity_no_promo, color='green')
plt.xlabel('Price')
plt.ylabel('Elasticity')
plt.title('Price Elasticity of Purchase Probability with and without Promotion');
```
